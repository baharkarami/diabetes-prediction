# **Diabetes Predictor**

## 1) Load All Models

```python
import joblib
import pandas as pd
import numpy as np

models = {
    "logistic_regression": joblib.load("../models/logistic_regression_model.pkl"),
    "random_forest": joblib.load("../models/random_forest_model.pkl"),
    "svm": joblib.load("../models/svm_model.pkl"),
    "gradient_boosting": joblib.load("../models/gradient_boosting_model.pkl"),
    "xgboost": joblib.load("../models/xgboost_model.pkl")
}
```

## 2) Get Inputs from User

```python
def get_user_input():
    print("Enter patient data:")

    pregnancies = float(input("Number of Times Pregnant: "))
    glucose = float(input("Plasma Glucose Concentration: "))
    blood_pressure = float(input("BloodPressure (mm Hg): "))
    skin_thickness = float(input("Triceps Skin Fold Thickness (mm): "))
    insulin = float(input("Insulin: "))
    bmi = float(input("BMI: "))
    dpf = float(input("DiabetesPedigreeFunction: "))
    age = float(input("Age: "))

    return pd.DataFrame({
        "Pregnancies": [pregnancies],
        "Glucose": [glucose],
        "BloodPressure": [blood_pressure],
        "SkinThickness": [skin_thickness],
        "Insulin": [insulin],
        "BMI": [bmi],
        "DiabetesPedigreeFunction": [dpf],
        "Age": [age]
    })
```

## 3) Select a Model

```python
print("Available models:")
for i, name in enumerate(models.keys()):
    print(f"{i}: {name}")

choice = input("Enter model index or type 'all': ")
```

## 4) Make Prediction

```python
user_data = get_user_input()

if choice == "all":
    print("\n--- Predictions ---")
    for name, model in models.items():
        pred = model.predict(user_data)[0]
        print(f"{name}: {'Diabetic' if pred==1 else 'Not Diabetic'}")

else:
    model_name = list(models.keys())[int(choice)]
    model = models[model_name]

    pred = model.predict(user_data)[0]
    print("\nPrediction:", "Diabetic" if pred==1 else "Not Diabetic")
```
